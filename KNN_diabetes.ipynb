<h3 style="margin-top:0;">1. Giới thiệu thuật toán K-NN (K Nearest Neighbors)</h3>

<p style="text-align:justify;">K-nearest neighbor là một trong những thuật toán supervised-learning đơn giản nhất trong Machine Learning. Khi training, thuật toán này không học một điều gì từ dữ liệu training, mọi tính toán được thực hiện khi nó cần dự đoán kết quả của dữ liệu mới. Thuật toán K-NN áp dụng trong bài toán trên thuộc lớp bài toán Classification (Phân Loại) của Machine Learing.  
Thuật toán KNN cho rằng những dữ liệu tương tự nhau sẽ tồn tại gần nhau trong một không gian, từ đó công việc của chúng ta là sẽ tìm k điểm gần với dữ liệu cần kiểm tra nhất. Việc tìm khoảng cách giữa 2 điểm củng có nhiều công thức có thể sử dụng, tùy trường hợp mà chúng ta lựa chọn cho phù hợp. Đối với phạm vi của đề tài nhóm thực hiện, nhóm sử dụng hàm Euclide để tính toán khoảng cách.<p>

<h3 style="margin-top:0;">Xây dựng hàm tính khoảng cách dựa trên hàm khoảng cách Euclide</h3>
<img src="https://www.delftstack.com/img/Math/euclidean%20distance.png"></img>

```python
def computeDistance(data1,data2):
  res = 0;
  for i in range(8):
    res += (data1[i] - data2[i])**2
  return math.sqrt(res)
```

<h3 style="margin-top:0;">Xây dựng hàm K-NN bao gồm việc tính khoảng cách và lấy nhãn của các điểm trong tập Training với điểm đang được xét</h3>

```python
def computeKNN(trainingData,point,k):
  distance = []#mảng chứa khoảng cách các điểm
  for data in trainingData:
    #khai bao dictionary bao gồm nhãn và khoảng cách từ điểm trong tập training đến điểm được xét
    label_distance = {
        "label" : data[-1],
        "dis": computeDistance(data,point)
    }
    distance.append(label_distance)
    #Sắp xếp lại mảng theo thứ tự tăng dần dựa trên khoảng cách
  distance.sort(key = lambda x: x["dis"])
  label = []
    #Lấy ra label của k điểm gần nhất đơi với điểm được xét
  for element in distance:
    label.append(element["label"])
  return label[:k]
```

<h3 style="margin-top:0;">Xây dựng hàm dự đoán kết quả dựa trên nhãn của k điểm gần nhất</h3>
<p>Nhãn sẽ được quyết định dựa trên <b>Major voting</b>, tức trong k điểm gần nhất nhãn nào xuất hiện nhiều hơn thì nhãn đó sẽ được gán cho điểm đang xét</p>

```python
def predict(label):
  num1 = 0
  num0 = 0
#Đếm số lượng mỗi nhãn trong k điểm gần nhất
  for element in label:
    if int(element) == 1:
      num1 = num1 + 1
    else:
      num0 = num0 + 1
#so sánh số lượng mỗi nhãn để chọn ra nhãn gán cho điểm
  if num1 > num0 :
    return 1
  else:
    return 0
```

<h3 style="margin-top:0;">Xây dựng hàm đánh giá kết quả giữa nhãn thật và nhãn được gán khi phân loại</h3>
<p>So sánh kết quả giữa 2 nhãn rồi phân loại vào 4 nhóm sau đây:</p>
<ul>
    <li>TP - True Positive: Nhãn thật thuộc nhóm dương và nhãn phân loại cũng thuộc nhóm dương</li>
    <li>FN - False Negative: Nhãn thật thuộc nhóm dương nhưng nhãn phân loại thuộc nhóm âm</li>
    <li>FP - False Positive: Nhãn thật thuộc nhóm âm nhưng nhãn phân loại thuộc nhóm dương</li>
    <li>TN - True Negative: Nhãn thật thuộc nhóm âm và nhãn phân loại cũng thuộc nhóm âm</li>
</ul>

```python
def evaluate(x,y):
    tp = 0
    fn = 0
    fp = 0
    tn = 0
    if (x==1 and y == 1):
        tp = 1
    elif (x == 1 and y == 0):
        fn = 1
    elif (y == 1):
        fp = 1
    else:
        tn = 1
    return [tp,fn,fp,tn]
```

<h3 style="margin-top:0;">Thực hiên quá trình phân loại dữ liệu trong tập Test và kết quả phân loại</h3>
<p style="font-size:16px;">Lựa chọn k = 23, tức 23 điểm gần điểm được xét nhất để gán nhãn cho điểm đó</p>

```python
k = 23
TP = 0
FN = 0  
FP = 0  
TN = 0  
truthVote=[]
predictionVote=[]
#Thực hiện phân loại dữ liệu trong TestData
for element in testData:
    label = computeKNN(trainingData, element, k)
    tp, fn, fp, tn = evaluate(int(element[-1]), predict(label))  
    truthVote.append(int(element[-1]))
    predictionVote.append(predict(label))
    TP += tp
    FN += fn
    FP += fp
    TN += tn
data_size = len(testData)
#Hiển thi số lượng các nhãn dán của 4 nhóm đánh giá chất lượng
print("TP =", TP) 
print("FN =", FN) 
print("FP =", FP) 
print("TN =", TN) 
#Tính toán các chỉ số để đánh giá độ chính xác của thuật toán
accuracy =round(float( (TP + TN)/data_size),10)
precision = float(TP/(TP+FP)) 
recall = float(TP/(TP+FN)) 

print(f"Accuracy: {accuracy*100:.2f}%")
print(f"Precision: {precision*100:.2f}%")
print(f"Recall: {recall*100:.2f}%")
```

<b>Chú thích các kết quả phân loại dữ liệu:</b>
<ul>
    <li><b>TP - True Positive:</b> Số người bị bệnh và được dự đoán là bị bệnh</li>
    <li><b>FN - False Negative:</b> Số người bị bệnh nhưng được dự đoán là không bị bệnh</li>
    <li><b>FP - False Positive:</b> Số người không bị bệnh những được dự đoán là bị bệnh</li>
    <li><b>TN - True Negative:</b> Số người không bị bệnh và được dự đoán là không bị bệnh</li>
    <li><b>Accuracy:</b> Tỉ lệ giữa những người được dự đoán đúng tình trạng và tổng số người trong tập dữ liệu Test</li>
    <li><b>Precision:</b> Tỉ lệ số người bị bệnh mô hình dự đoán đúng trên tổng số người mô hình dự đoán là bị bệnh</li>
    <li><b>Recall:</b> Tỉ lệ số người bị bệnh mô hình dự đoán đúng trên tổng số người thật sự bị bệnh (hay tổng số người được gán nhãn là bị bệnh ban đầu).</li>
</ul>

Tuy nhiên đây vẫn chưa phải con số mong muốn nên ta tiếp tục đến với các phương pháp tối ưu thuật toán</p>

<h2 style="margin-top:0;">2. Tối ưu thuật toán KNN với phương pháp đánh trọng số cho các điểm (WEIGHTED KNN)</h3>

<p style="text-align: justify;">Trong kỹ thuật Major voting bên trên, mỗi điểm trong k điểm gần nhất được coi là có vai trò như nhau và giá trị lá phiếu của mỗi điểm này là như nhau. Tuy nhiên điều này có thể gây nên sự không công bằng, vì rõ ràng rằng những điểm gần hơn nên có trọng số cao hơn (càng thân cận thì càng tin tưởng) và ngược lại càng điểm ở xa hơn thì có trọng số nhỏ hơn. Vậy nên ta sẽ đánh trọng số khác nhau cho mỗi điểm trong k điểm gần nhất này. Cách đánh trọng số phải thoải mãn điều kiện là một điểm càng gần điểm test data thì phải được đánh trọng số càng cao (tin tưởng hơn). 
   <br>
<br>Có rất nhiều cách để đánh trọng số cho dữ liệu trong thực tế, trong phương pháp này nhóm lựa chọn đánh <b>trọng số</b> cho các điểm bằng cách lấy <b>nghịch đảo của khoảng cách từ điểm đó đến điểm được Test</b>. Dễ thấy các điểm càng gần điểm Test thì khoảng cách càng nhỏ dẫn đến trọng số bằng nghịch đảo khoảng cách sẽ càng lớn. Ở bài toán đang xét, 768 điểm dữ liệu hoàn toàn riêng biệt nên không có trường hợp khoảng cách giữa điểm dữ liệu đang xét với điểm dữ liệu trong tập huấn luyện bằng 0. Vì vậy, ta hoàn toàn có thể đánh trọng số bằng cách lấy nghịch đảo khoảng cách</p>

<h3 style="margin-top:0;">Xây dựng hàm K-NN bao gồm việc tính khoảng cách và lấy nhãn của các điểm trong tập Training với điểm đang được xét</h3>

```python
def new_computeKNN(trainingData,point,k):
  distance = []
  for data in trainingData:
    #khai bao dictionary
    label_distance = {
        "label" : data[-1],
        "dis": computeDistance(data,point)
    }
    distance.append(label_distance)
  distance.sort(key = lambda x: x["dis"])

  return distance[:k]
```

<h3 style="margin-top:0;">Xây dựng hàm dự đoán kết quả mới dựa trên tổng trọng số của mỗi nhãn</h3>
<p style="text-align: justify;">Thiết lập 2 tổng weight1 cho nhãn 1 và weight0 cho nhãn 0, cứ mỗi điểm trong k điểm gần nhất ta vừa xét thuộc nhãn nào thì trọng số của điểm đó sẽ được cộng thêm vào tổng tương ứng. Nhãn nào có tổng trọng số lớn hơn thì sẽ được chọn gán cho điểm Test</p>
Trọng số = $\frac{1}{distance}$

```python
def new_predict(label):
  weight1 = 0
  weight0 = 0
  for element in label:
    if int(element["label"]) == 1:
      weight1 += 1/float(element["dis"])#tính tổng trọng số cho label 1
    else:
      weight0 += 1/float(element["dis"])#tính tổng trọng số cho label 0
  if weight1 > weight0: 
    return 1
  else:
    return 0
```

<p style="font-size:16px;">Ta vẫn lựa chọn k = 23, tức 23 điểm gần điểm được xét nhất để gán nhãn cho điểm đó giống với phương pháp Major Voting để tiện so sánh</p>

```python
k = 23
TP = 0
FN = 0
FP = 0
TN = 0
truth=[]
prediction=[]
for element in testData:
    label = new_computeKNN(trainingData, element, k)
    truth.append(int(element[-1]))
    prediction.append(new_predict(label))
    tp, fn, fp, tn = evaluate(int(element[-1]), new_predict(label))
    TP += tp
    FN += fn
    FP += fp
    TN += tn
print("TP =", TP) 
print("FN =", FN) 
print("FP =", FP) 
print("TN =", TN) 

accuracy =round(float( (TP + TN)/data_size),10)
precision = float(TP/(TP+FP)) 
recall = float(TP/(TP+FN))

print(f"Accuracy: {accuracy*100:.2f}%")
print(f"Precision: {precision*100:.2f}%")
print(f"Recall: {recall*100:.2f}%")
```

**LÝ DO CẦN CHUẨN HÓA** 
<p style="text-align: justify;">Khi có một thuộc tính trong dữ liệu (hay phần tử trong vector) lớn hơn hoặc bé hơn các thuộc tính khác rất nhiều (ví dụ thay vì đo bằng m thì một kết quả lại tính bằng mm), khoảng cách giữa các điểm sẽ bị phụ thuộc rất lớn vào thuộc tính này. Dẫn đến các thuộc tính khác sẽ bị mờ nhạt đi trong việc tính toán khoảng cách gây nên sự nhiễu dự liệu. Để có được kết quả chính xác hơn, một kỹ thuật thường được dùng là Data Normalization (chuẩn hóa dữ liệu) để đưa các thuộc tính có đơn vị đo khác nhau về cùng một khoảng giá trị, thường là từ 0 đến 1, trước khi thực hiện KNN.</p>

Ví dụ: Trường DiabetesPedigreeFunction trong dataset có giá trị [0.078, 2.42]

Nhưng khi tính khoảng cách giữa các điểm dữ liểu bằng khoảng cách euclid thì nó không đóng nhiều ảnh hưởng cho việc phân loại dữ liệu do miền giá trị của nó quá bé so với miền giá trị của các thuộc tính còn lại.

**PHƯƠNG PHÁP CHUẨN HÓA** 
Ta chuẩn hóa theo công thức  $\frac
{X_i-MIN_i}{MAX_i-MIN_i}$ 
Để đữa các dữ liệu về miền giá trị [0,1]




<h4 style="margin-top:0;">Tìm max, min từng thuộc tính</h4>

```python
max_item = np.array(filtered_dataset[0][:8])
min_item = np.array(filtered_dataset[0][:8])
for item in filtered_dataset:
    for i in range(8):
            if item[i] > max_item[i]:
                max_item[i] = item[i]
            if item[i] < min_item[i]:
                min_item[i] = item[i]
print(max_item)
print(min_item)
max_item = max_item.astype(float)
min_item = min_item.astype(float)
```

**Chuẩn hóa dữ liệu dựa trên công thức**

```python
for item in filtered_dataset:
    for i in range(8):
        item_value = float(item[i])
        item[i] = float((item_value - min_item[i])/(max_item[i] - min_item[i]))
```

**In ra các giá trị của tập dữ liệu sau khi được chuẩn hóa. Các giá trị mới đã thỏa mãn nằm trong miền [0,1]**

```python
def kAccuracyCalculate(testData,trainingData,computeKNN,predict):
    data_points=[]
    data_size=len(testData)
    for k in range(5,100):
        TP = 0
        FN = 0
        FP = 0
        TN = 0

        for element in testData:
            label = computeKNN(trainingData, element, k)
            tp, fn, fp, tn = evaluate(int(element[-1]), predict(label))
            TP += tp
            FN += fn
            FP += fp
            TN += tn
        accuracy =round(float( (TP + TN)/data_size),10)*100
        data_points.append([k,accuracy])
    return data_points
```

<h3 style="margin-top:0;">Biểu đồ đánh giá tỉ lệ Error Rate (tỉ lệ dự đoán sai) theo giá trị của k từ 5 đến 100 của từng phương pháp</h3>

```python
def kErrorCalculate(testData,trainingData,computeKNN,predict):
    Edata_points=[]
    data_size=len(testData)
    for k in range(5,100):
        TP = 0
        FN = 0
        FP = 0
        TN = 0

        for element in testData:
            label = computeKNN(trainingData, element, k)
            tp, fn, fp, tn = evaluate(int(element[-1]), predict(label))
            TP += tp
            FN += fn
            FP += fp
            TN += tn
        errorRate =round(float( (FP + FN)/data_size),10)*100
        Edata_points.append([k,errorRate])
    return Edata_points
```

```python
Edata_pointsWeight=kErrorCalculate(testData,trainingData,new_computeKNN,new_predict)
Edata_pointsVote=kErrorCalculate(testData,trainingData,computeKNN,predict)
Edata_pointsNormalize=kErrorCalculate(filtered_testData,filtered_trainingData,new_computeKNN,new_predict)
```

```python
import matplotlib.pyplot as plt

x_values, y_values = zip(*Edata_pointsWeight)
x_valuesVote, y_valuesVote = zip(*Edata_pointsVote)
x_valuesN, y_valuesN = zip(*Edata_pointsNormalize)

plt.figure(figsize=(10,6))

plt.plot(x_values, y_values, marker='o',linestyle='-',markersize=2,linewidth=1,color='b',label='Weighted KNN')
plt.plot(x_valuesVote, y_valuesVote, marker='o',linestyle='-',markersize=2,linewidth=1,color='r',label='Major Vote')
plt.plot(x_valuesN, y_valuesN, marker='o',linestyle='-',markersize=2,linewidth=1,color='g',label='Normalize Data')


min_point = min(Edata_pointsWeight, key=lambda point: point[1])
min_pointVote = min(Edata_pointsVote, key=lambda point: point[1])
min_pointN = min(Edata_pointsNormalize, key=lambda point: point[1])

plt.scatter([min_point[0]], [min_point[1]], color='b', label=f'Min Weighted KNN: {format(min_point[1],".2f")}%', zorder=5)
plt.scatter([min_pointVote[0]], [min_pointVote[1]], color='r', label=f'Min Major Vote: {format(min_pointVote[1],".2f")}%', zorder=5)
plt.scatter([min_pointN[0]], [min_pointN[1]], color='g', label=f'Min Normalize Data: {format(min_pointN[1],".2f")}%', zorder=5)
plt.xlabel('Gía trị k - k điểm gần nhất để lựa chọn nhãn')
plt.ylabel('Tỉ lệ dự đoán sai - Error Rate (%)')

plt.title('Biểu đồ Error Rate theo giá trị k')

plt.legend()
plt.show()
```
